```python
# Step 2: Data Cleaning

match_df = pd.read_csv("User Reviews.csv")

# Fill missing values with mode
for column in match_df.columns:
    if not match_df[column].mode().empty:
        match_df[column] = match_df[column].fillna(
            match_df[column].mode()[0]
        )

# Remove duplicates
match_df.drop_duplicates(inplace=True)

# Convert Sentiment_Polarity and Sentiment_Subjectivity to numeric
match_df["Sentiment_Polarity"] = pd.to_numeric(
    match_df["Sentiment_Polarity"],
    errors="coerce"
)

match_df["Sentiment_Subjectivity"] = pd.to_numeric(
    match_df["Sentiment_Subjectivity"],
    errors="coerce"
)

# Remove rows with missing sentiment values
match_df.dropna(
    subset=["Sentiment_Polarity", "Sentiment_Subjectivity"],
    inplace=True
)

print(match_df.info())
print(match_df.head())
```

```python
def power_group(installs):
    if installs >= 10000000:
        return "Very Popular"
    elif installs >= 1000000:
        return "Popular"
    elif installs >= 100000:
        return "Average"
    else:
        return "Less Popular"

# Clean Installs column
apps_df["Installs"] = (
    apps_df["Installs"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace("+", "", regex=False)
)

apps_df["Installs"] = pd.to_numeric(
    apps_df["Installs"],
    errors="coerce"
)

apps_df["power_group"] = apps_df["Installs"].apply(power_group)

apps_df[["App", "Installs", "power_group"]].head()
```

```python
import os
import plotly.io as pio

plot_containers = ""

def save_plot_as_html(fig, filename, insight):
    global plot_containers

    filepath = os.path.join(html_files_path, filename)

    html_content = pio.to_html(
        fig,
        full_html=False,
        include_plotlyjs='inline'
    )

    plot_containers += f"""
    <div class="plot-container" id="{filename}" onclick="openPlot('{filename}')">
        <div class="plot">{html_content}</div>
        <div class="insights">{insight}</div>
    </div>
    """

    fig.write_html(
        filepath,
        full_html=False,
        include_plotlyjs='inline'
    )
```

```python
# Figure 1
category_counts = apps_df["Category"].value_counts().nlargest(20)

fig1 = px.bar(
    x=category_counts.index,
    y=category_counts.values,
    labels={"x": "Category", "y": "Number of Apps"},
    title="Top App Categories on Google Play Store",
    color=category_counts.index,
    color_discrete_sequence=px.colors.sequential.Plasma,
    width=800,
    height=500
)

fig1.update_layout(
    plot_bgcolor="black",
    paper_bgcolor="black",
    font_color="white",
    title_font={"size": 16},
    xaxis=dict(title_font={"size": 12}),
    yaxis=dict(title_font={"size": 12}),
    margin=dict(l=50, r=50, t=80, b=80)
)

fig1.update_traces(
    marker=dict(
        line=dict(color="white", width=1)
    )
)

save_plot_as_html(
    fig1,
    "Category Graph 1.html",
    "The top categories on the Play Store are dominated by tools, entertainment, and productivity apps."
)
```

```python
# Figure 2: Type Analysis Graph

type_counts = apps_df["Type"].value_counts()

fig1 = px.pie(
    values=type_counts.values,
    names=type_counts.index,
    title="App Type Distribution",
    color_discrete_sequence=px.colors.sequential.RdBu,
    width=800,
    height=500
)

fig1.update_layout(
    plot_bgcolor="black",
    paper_bgcolor="black",
    font_color="white",
    title_font={"size": 16},
    margin=dict(l=50, r=50, t=80, b=80)
)

fig1.update_traces(
    marker=dict(
        line=dict(color="white", width=1)
    )
)

save_plot_as_html(
    fig1,
    "Type Graph 2.html",
    "Most apps on the Google Play Store are free, while a smaller percentage are paid apps."
)
```
